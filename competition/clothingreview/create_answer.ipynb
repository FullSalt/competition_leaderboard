```python
import pandas as pd
```

```python
#  元データの読み込み
df_origin = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
df_origin.head()
```

```python
df_origin.shape
```

```python
df_origin.rename(columns={'Review Text': 'Review'}, inplace=True)
df_origin.head()
```

```python
# 受講生に配布するtestデータの読み込み
df_test = pd.read_csv('test_Day3.csv')
df_test.head()
```

```python
df_test.shape
```

```python
df_train = pd.read_csv('train_Day3.csv')
df_train.head()
```

```python
df_origin.columns, df_test.columns
```

```python
# df_origin と df_test で共通のカラムを抽出
common_columns = list(set(df_origin.columns) & set(df_test.columns))
common_columns
```

```python
df_test_with_target = df_origin.merge(df_test, on=common_columns, how='right')
df_test_with_target.head()
```

```python
df_test_with_target.tail()
```

```python
df_test_with_target.shape
```

```python
df_test_with_target['ID'].value_counts()
```

```python
df_test_with_target[df_test_with_target['ID'] == 87]
```

```python
df_test_with_target = df_test_with_target[df_train.columns]
df_test_with_target.head()
```

```python
# 重複業の削除
df_test_with_target.drop_duplicates(inplace=True)
df_test_with_target.shape
```

```python
df_test_with_target.to_csv('test_Day3_with_target.csv', index=False)
```

```python
df_test_answer = df_test_with_target[['Recommended IND']]
df_test_answer.to_csv('submission_answer.csv', index=False)
```
